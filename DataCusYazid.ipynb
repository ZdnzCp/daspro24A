<h1>Data Penjualan HandPhone</h1>

```python
import numpy as np
import pandas as pd

n_data = 1000

np.random.seed(42)

usia = np.random.randint(18, 61, size=n_data)
gender = np.random.randint(0, 2, size=n_data)
transaksi = np.random.randint(0, 51, size=n_data)
kategori = np.random.choice(['Buruk', 'Sedang', 'Bagus'], size=n_data)
tipe_hp = np.random.choice(['Low-End', 'Mid-Range', 'High-End'], size=n_data)
harga = np.random.randint(1000000, 5000001, size=n_data)
rating = np.round(np.random.uniform(1.0, 5.0, size=n_data), 1)
garansi = np.random.choice([0, 1], size=n_data)
durasi_penggunaan = np.random.randint(1, 37, size=n_data)
respon_servis = np.random.choice(['Cepat', 'Lambat', 'Normal'], size=n_data)

df = pd.DataFrame({
    'Usia': usia,
    'Gender': gender,
    'Transaksi': transaksi,
    'Kategori': kategori,
    'TipeHP': tipe_hp,
    'Harga': harga,
    'Rating': rating,
    'Garansi': garansi,
    'DurasiPakai': durasi_penggunaan,
    'ResponServis': respon_servis
})

print(df)
print(df.info())
print("-----------------------------------------------")
print(df.head())
df.to_csv("dataPenjualanHP.csv", index=False)
print("-----------------------------------------------")

print(f"Usia Minimum: {df.Usia.min()}")
print(f"Usia Maksimum: {df.Usia.max()}")

jmlL = df[df.Gender == 1]
print(f"Jumlah Customer Laki-Laki: {jmlL.shape[0]}")

jmlP = df[df.Gender == 0]
print(f"Jumlah Customer Perempuan: {jmlP.shape[0]}")

print(f"Rata-rata Transaksi per Bulan: {df.Transaksi.mean():.2f}")
print(f"Rata-rata Harga HP: Rp{df.Harga.mean():,.0f}")
print(f"Rata-rata Rating Kepuasan: {df.Rating.mean():.2f}")
print(f"Persentase Bergaransi: {df[df.Garansi == 1].shape[0] / n_data * 100:.2f}%")
```
